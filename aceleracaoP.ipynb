Aceleração Próton

```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes
c = 299792458  # velocidade da luz em m/s
q = 1.60217662e-19  # carga elementar em coulombs
V = 1e9  # voltagem em volts
m = 1.6726219e-27  # massa do próton em kg

# Cálculo da energia máxima
Emax = q * V

# Energia cinética máxima
Kmax = Emax - m * c**2

# Intervalo de energias cinéticas
K = np.linspace(0, Kmax, 1000)

# Velocidade relativa
v = np.sqrt(2 * K / m)

# Tamanho do objeto
R = 1  # Em metros

# Campo magnético
B = np.linspace(0, 10, 1000)  # Em Tesla

# Cálculo do gráfico de Hillas
fonte = R * B * v / c

# Plotar o gráfico
plt.figure(figsize=(10, 6))
plt.plot(B, fonte)
plt.xlabel('Campo Magnético (T)')
plt.ylabel('Tamanho da Fonte (m)')
plt.title('Gráfico de Hillas para Aceleração de um Próton até Emax')
plt.grid(True)
plt.show()
```
